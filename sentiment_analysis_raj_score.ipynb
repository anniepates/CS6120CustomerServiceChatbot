```python
data = pd.read_csv("amazon_reviews_us_Digital_Software_v1_00.tsv", sep='\t')
data = data [:10000]
```

```python
rev_class = np.zeros
data["reviw_class"] = rev_class
data = data.drop(columns = ['marketplace','customer_id','review_id','product_id','product_parent',
                                         'product_title','product_category','vine','review_date','review_headline'
                                         ])
data
```

```python
for index, row in data.iterrows():
    if row['verified_purchase'] == 'Y':
        data.at[index, 'verified_purchase'] = 1
    elif row['verified_purchase'] == 'N':
        data.at[index, 'verified_purchase'] = 0

    if row['star_rating'] == 5 or row['star_rating'] == 4:
        data.at[index, 'reviw_class'] = 1
    else:
        data.at[index, 'reviw_class'] = 0
```

```python
data
```

```python
split = int(0.8*len(data))
data_train, data_test = data[:split], data[split:]
```

```python
X_train = data_train.drop(columns = ['reviw_class'])
X_test = data_test.drop(columns = ['reviw_class'])


# X_train = data_train['review_body']
# X_test = data_test['review_body']
```

```python
VOCAB_SIZE = 1000
encoder = tf.keras.layers.TextVectorization(
    max_tokens=VOCAB_SIZE
)

# Fit the encoder to the training text data
# encoder.adapt(X_train['review_body'].values)



# encoder.adapt(X_train)
```

```python
# model = tf.keras.Sequential([
#     encoder,
#     tf.keras.layers.Embedding(
#         input_dim=len(encoder.get_vocabulary()),
#         output_dim=64,
#         # Use masking to handle the variable sequence lengths
#         mask_zero=True),
#     tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(100)),
#     tf.keras.layers.Dense(100, activation='relu'),
#     tf.keras.layers.Dense(50, activation='relu'),
#     tf.keras.layers.Dense(1)
# ])
```

```python
model = tf.keras.Sequential([
    encoder,
    tf.keras.layers.Embedding(
        input_dim=len(encoder.get_vocabulary()),
        output_dim=128,  # Increased output dimension
        mask_zero=True),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(256, return_sequences=True)),  # Return sequences for deeper layers
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(128)),  # Reduced units for deeper layers
    tf.keras.layers.Dense(256, activation='relu'),  # Increased units
    tf.keras.layers.Dropout(0.5),  # Dropout layer for regularization
    tf.keras.layers.Dense(128, activation='relu'),  # Additional dense layer
    tf.keras.layers.Dense(64, activation='relu'),  # Additional dense layer
    tf.keras.layers.Dense(1)
])
```

```python
## TEST
import h5py
# Path to your .h5 file
file_path = 'my_model_weights.h5'

with h5py.File(file_path, 'r') as f:
    # Iterate through each group (corresponding to a layer)
    for layer_name in f.keys():
        group = f[layer_name]
        # Retrieve each dataset within the group using correct key access
        # Create a list to store weights for the current layer
        weights = []
        for dset_name in group.keys():
            # Access the dataset and read its data
            dataset = group[dset_name]
            data = dataset[:]  # Correct way to access data from the dataset
            weights.append(data)
        # Set the weights to the corresponding layer in the model
        layer = model.get_layer(layer_name)
        layer.set_weights(weights)
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```
